```python
import requests
from bs4 import BeautifulSoup
from lxml import html
```

```python
# Base URL for making relative links absolute
base_url = 'https://www.stedin.net'

# URL of the FAQ page
url = f'{base_url}/klantenservice/veelgestelde-vragen/aansluiten-en-afsluiten'

# Headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

# Request the main FAQ page
response = requests.get(url, headers=headers)
response.raise_for_status()  # Ensure the request was successful

# Parse the main page HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <a> tags with a 'title' attribute, which seem to represent the questions
question_links = soup.find_all('a', title=True)

# Iterate through each <a> tag, follow the link, and extract the answer
for link in question_links:
    question_title = link['title']
    answer_url = link['href']
    
    # Make sure the URL is absolute
    if not answer_url.startswith('http'):
        answer_url = base_url + answer_url
    
    # Fetch the page with the answer
    answer_response = requests.get(answer_url, headers=headers)
    answer_response.raise_for_status()
    answer_soup = BeautifulSoup(answer_response.text, 'html.parser')
    
    # Extract all the text from the answer div (modify the selector to fit the answer structure)
    answer_divs = answer_soup.find_all('div', class_='col-12')
    
    # Loop through the divs to find the specific answer by checking the presence of the question in the h3 tag
    for div in answer_divs:
        h3_tag = div.find('h3')
        if h3_tag and h3_tag.get_text(strip=True) == question_title:
            # Extract all text content within this div
            answer_text = div.get_text(separator=' ', strip=True)
            print(f'Question: {question_title}')
            print(f'Answer: {answer_text}\n')
            break
```
